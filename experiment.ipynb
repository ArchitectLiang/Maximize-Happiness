```python
res = pd.DataFrame(columns = ['s1', 's2','ha', 'stress']) 
res
```

```python
ss = []
for i in range(50):
    for j in range(i + 1,50):
        row = dic[i]
        col = dic[j]
        if row > col:
            happyness = df[j][i]
            stress = df[i][j]
        else:
            happyness = df[i][j]
            stress = df[j][i]
            
        res = res.append({'s1': int(i), 's2' : int(j), 'ha':happyness, 'stress': stress}, ignore_index=True)
```

```python
# res.astype({'s1': 'int32', 's2': 'int32'}).dtypes
res["s1"] = res["s1"].astype(int)
res["s2"] = res["s2"].astype(int)

res.head()
```
